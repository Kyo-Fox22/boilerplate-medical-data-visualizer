```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1
df = pd.read_csv('medical_examination.csv')
df.head()
```

```python
# 2
df['overweight'] = ((df['weight']/(df['height']*0.01)**2) > 25).astype(int)
df.head()
```

```python
# 3
df['cholesterol'] = df['cholesterol'].map(lambda x: 0 if x == 1 else 1)
df['gluc'] = df['gluc'].map(lambda x: 0 if x == 1 else 1)
df.head()
```

```python
# 4 involves steps 5 to 9
# 5
df_cat = df.melt(value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'], id_vars=['cardio'])
df_cat.head()
```

```python
# 6
df_cat = df_cat.groupby('cardio').value_counts()
df_cat
```

```python
#7
sns.catplot(df_cat.reset_index(name = 'count'), x='variable', y='count', hue='value', col='cardio', kind='bar')
```

```python
# 8
fig = sns.catplot(df_cat.reset_index(name = 'count'), x='variable', y='count', hue='value', col='cardio', kind='bar').figure
```

```python
df['height'].quantile(0.025)
```

```python
# 10 Consist of steps 11-16
# 11
df_heat = df[
    (df['ap_lo'] <= df['ap_hi']) &
    (df['height'] >= df['height'].quantile(0.025)) &
    (df['height'] <= df['height'].quantile(0.975)) &
    (df['weight'] >= df['weight'].quantile(0.025)) &
    (df['weight'] <= df['weight'].quantile(0.975))
]
```

```python
# 12
corr = df_heat.corr()
corr
```

```python
# 13
mask = np.triu((np.ones(corr.shape))).astype(bool)
mask
```

```python
# 14
fig, ax = plt.subplots(figsize=(12, 10))
```

```python
# 15
sns.heatmap(corr, mask = mask, annot = True, fmt = '.2f', ax = ax)
fig
```
